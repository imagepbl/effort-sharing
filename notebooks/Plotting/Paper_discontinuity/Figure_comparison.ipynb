```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
xr_dataread = xr.open_dataset('K:/Data/Data_effortsharing/DataUpdate_ongoing/xr_dataread.nc')
xr_combs = xr.open_dataset('K:/Data/Data_effortsharing/DataUpdate_ongoing/xr_combinedapproaches.nc')
```

```python
reg = 'IND'
fig, axes = plt.subplots(3, 3, figsize=(20, 10))
plt.subplots_adjust(hspace=0.3, wspace=0.3)
for ax_i, ax in enumerate(axes.flatten()):
    reg = ['USA', 'CHN', 'IND', 'RUS', 'ZAF', 'UGA', 'AUS', 'BRA', 'JPN'][ax_i]
    ax.set_xlim([1990, 2100])
    ax.plot(xr_dataread.Time, xr_dataread.GHG_hist_excl.sel(Region=reg), label='Historical emissions', c='k')
    ax.set_title(reg)
    for var_i, var in enumerate(['Approach1_gdp', 'Approach1_hdi', 'Approach2', 'Approach2t', 'GF']):
        ax.plot(xr_combs.Time, xr_combs.sel(Discount_factor=1.6, Historical_startyear=1990, Temperature=1.6, Risk=0.5, Scenario='SSP2', Convergence_year=2040, Region=reg)[var], label=var, c=['orange', 'purple', 'forestgreen', 'tomato', 'navy'][var_i])
plt.legend()
```
